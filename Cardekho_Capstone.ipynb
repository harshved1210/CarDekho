```python
# df['new_car_detail']:

rows = []   # Creates an empty list to store processed car details.

for data_str in df['new_car_detail']:   # Loops through each entry (string) in the DataFrame column.
    # Convert string to dictionary
    data_dict = ast.literal_eval(data_str)

    # Extract nested 'trendingText' information
    trending_info = data_dict.pop('trendingText', {})
    data_dict['trending_heading'] = trending_info.get('heading')
    data_dict['trending_desc'] = trending_info.get('desc')
    data_dict['trending_imgUrl'] = trending_info.get('imgUrl')

    # Append the processed dictionary to the rows list
    rows.append(data_dict)

# Create DataFrame
dv = pd.DataFrame(rows)
```

```python
rows = []
for data_str in df['new_car_overview']:
    # Convert string to dictionary
    data_dict = ast.literal_eval(data_str)

    # Extract and remove nested 'top' list
    top_list = data_dict.pop('top', [])
    # Extract key-value pairs from 'top' and add to data_dict
    for item in top_list:
        key = item.get('key')
        value = item.get('value')
        if key:  # Ensure key is valid
            data_dict[key] = value

    # Remove 'bottomData' (optional: process it if needed)
    data_dict.pop('bottomData', None)

    # Append processed dictionary to rows
    rows.append(data_dict)

# Create DataFrame
eg = pd.DataFrame(rows)

# Display the DataFrame
eg
```

**new_car_feature**

```python
df['new_car_feature'].iloc[0]
```

```python

rows = []

for data_str in df['new_car_feature']:
    # Convert string to dictionary
    data_dict = ast.literal_eval(data_str)

    # Initialize row dictionary
    row = {}

    # Extract top-level heading
    row["Heading"] = data_dict.get("heading", "")

    # Extract 'top' values as individual features
    if "top" in data_dict:
        for item in data_dict["top"]:
            value = item.get("value")
            if value:
                row[value] = True  # Mark feature as present

    # Extract features from 'data' section
    if "data" in data_dict:
        for section in data_dict["data"]:
            heading = section.get("heading", "")
            feature_list = [item.get("value") for item in section.get("list", []) if item.get("value")]
            if heading:
                row[heading] = ", ".join(feature_list)  # Store features as comma-separated strings

    # Append row data
    rows.append(row)

# Create DataFrame
qw = pd.DataFrame(rows)

# Fill missing values with False (or empty string if preferred)
qw.fillna(False, inplace=True)

# Display DataFrame
```

```python
row = []
for data_str in df['new_car_specs']:
  data_dict = ast.literal_eval(data_str)
  # dict_keys(['heading', 'top', 'data', 'commonIcon'])
  # Extracing top specsification
  top_spec = data_dict.pop('top',[])
  for item in top_spec:
    keys = item.get('key')
    value = item.get('value')
    if key:
      data_dict[keys] = value
  data_cat = data_dict.pop('data',[])
  for category in data_cat:
    subHeading = category.get('subHeading')
    lis = category.get('list')
    for l in lis:
      key = l.get('key')
      value = l.get('value')
      if key:
        head = f"{subHeading} - {key}"
        data_dict[head] = value

  row.append(data_dict)

jn = pd.DataFrame(row)
jn.head(3)
```
